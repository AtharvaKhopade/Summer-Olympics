
# **Summer Olympics Data Analysis Assignment**

```python
import pandas as pd                          #Importing pandas package          
df = pd.read_csv('summer.csv')               #Reading csv file 
```

### **1. In how many cities Summer Olympics is held so far?**

```python
a= len(df['City'].unique())
b= df['City'].unique()
print("Total Number of cities are :",a,"\n\n","Cities :",b)
```

### **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
df[df['Medal']=="Gold"].Sport.value_counts().head().plot(xlabel = 'Sport', ylabel = 'Gold', kind = 'bar', figsize = (8,5), color = 'lightblue')
```

### **3. Which sport is having most number of medals so far? (Top 5)**

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot( xlabel = 'Sport', ylabel = 'Medal',kind = "bar", figsize=(12, 5), color = 'lightblue')
```

### **4. Which player has won most number of medals? (Top 5)**

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot(xlabel = 'Athlete', ylabel = 'Medal', kind = 'bar',figsize=(12, 5), color = 'lightblue')
```

### **5. Which player has won most number Gold Medals of medals? (Top 5)**

```python
gd= df[df['Medal']=='Gold']
gd.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot( xlabel = 'Athlete', ylabel = 'Medal', kind = 'bar', color = 'lightblue',figsize=(12, 4))
```

### **6. In which year India won first Gold Medal in Summer Olympics?**

```python
 a= gd[gd['Country']=='IND']['Year'].min()
 print('India won first gold medal in summer olympics was :',a)
```

### **7. Which event is most popular in terms on number of players? (Top 5)**

```python
df.groupby("Event").count()['Athlete'].sort_values(ascending=False).head().plot( xlabel = 'Event', ylabel = 'Athlete', kind = 'bar', color = 'lightblue',figsize=(12, 5))
```

### **8. Which sport is having most female Gold Medalists? (Top 5)**

```python
gold_medal_Female = gd[gd['Gender']=='Women']
data=[]
for sport in gold_medal_Female['Sport'].unique():
  data.append([sport, len(gold_medal_Female[gold_medal_Female['Sport']==sport])])

data = pd.DataFrame(data, columns=['Sport', 'gold_medal_Female']).sort_values(by='gold_medal_Female', ascending=False).head().plot(x='Sport', y='gold_medal_Female', kind='bar', color ='lightblue', figsize=(12, 5))
```
